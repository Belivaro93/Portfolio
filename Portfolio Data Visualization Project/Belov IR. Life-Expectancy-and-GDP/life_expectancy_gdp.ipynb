# Introduction 

## Import the Python modules

```python
#Загрузка необходимых библиотек
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

## Load the data

```python
#Загрузка данных
df = pd.read_csv('all_data.csv')
print(df.head())
```

# Explore the data

```python
#Исследование значений полей в DateFrame
print(df['Country'].unique())
print(df['Year'].unique())
```

```python
#Переименование полей с длинными названиями для удобства обращения к ним
df = df.rename({'Life expectancy at birth (years)':'LEAB'},axis = "columns")
df['GDP'] = df['GDP']/(10**9)
print(df.head())
```

## LEAB and GDP distributions per years

```python
#Распределение значений ожидаемой продолжительности жизни по годам
plt.figure(figsize = (15,10))
sns.boxplot(data = df, x = 'Year', y = 'LEAB')
plt.title('Life Expentancies over World per Year')
plt.show()
```

```python
#Распределение значений ВВП в миллиардах долларов по годам
plt.figure(figsize = (15,10))
sns.boxplot(data = df, x = 'Year', y = 'GDP')
plt.title('GDP in billions dollars per year')
plt.show()
```

## LEAB and GDP over the years in 6 countries

```python
#Изменение ожидаемой продолжительности жизни по годам для 6 стран
plt.figure(figsize = (15,10))
sns.lineplot(data = df, x = 'Year', y = 'LEAB', hue = "Country")
plt.title('Life Expentancies in different countries per years')
plt.show()
```

```python
#Изменение ВВП по годам для 6 стран
plt.figure(figsize = (15,10))
sns.lineplot(data = df, x = 'Year', y = 'GDP', hue = "Country")
plt.title('GDP in different countries per years')
plt.show()
```

## Mean and Medians of LEAB and GDP in 6 countries

```python
#Средние и медианные значения ожидаемой продолжительности жизни в 6 странах с 2000 по 2015 годы
leab_means = df.groupby('Country').LEAB.mean().reset_index()
leab_medians = df.groupby('Country').LEAB.median().reset_index()

plt.figure(figsize = (15,10))
plt.subplot(2,1,1)
sns.barplot(data = leab_means, y = 'Country', x = 'LEAB')
plt.title('Means of Life Expentancies in different countries')
plt.subplot(2,1,2)
sns.barplot(data = leab_medians, y = 'Country', x = 'LEAB')
plt.title('Medians of Life Expentancies in different countries')
plt.show()
```

```python
#Средние и медианные значения ВВП в 6 странах с 2000 по 2015 годы
gdp_means = df.groupby('Country').GDP.mean().reset_index()
gdp_medians = df.groupby('Country').GDP.median().reset_index()

plt.figure(figsize = (15,10))
plt.subplot(2,1,1)
sns.barplot(data = gdp_means, y = 'Country', x = 'GDP')
plt.title('Means of GDP in different countries')
plt.subplot(2,1,2)
sns.barplot(data = gdp_medians, y = 'Country', x = 'GDP')
plt.title('Medians of GDP in different countries')
plt.show()
```

## Relationship between LEAB and GDP

```python
#Зависимость ожидаемой продолжительности жизни от уровня ВВП в 6 странах с 2000 по 2015 годы
plt.figure(figsize=(15,10))
sns.scatterplot(data = df, x = 'GDP', y = 'Life expectancy at birth (years)', hue = 'Country')
plt.show()
```
